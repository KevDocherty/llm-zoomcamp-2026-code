# A compact RAG assistant

This notebook assembles the reusable ingestion and RAG helpers into a small course FAQ assistant. It demonstrates the default assistant, broad and specific questions, and stricter grounding instructions.

## 1. Configure the client and retrieval index

Load environment variables, create the OpenAI client, load the FAQ documents, and build the in-memory search index.

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from openai import OpenAI
openai_client = OpenAI()
```

```python
from rag_helper import RAGBase
from ingest import load_faq_data, build_index
```

```python
documents = load_faq_data()
index = build_index(documents)
assistant = RAGBase(
    index=index, 
    llm_client=openai_client
)
```

## 2. Test the default assistant

Start with an enrollment question and then try a broad, underspecified query. Broad queries such as `docker` may retrieve several unrelated FAQ entries, illustrating why specific questions generally produce better answers.

```python
assistant.rag("I just discovered the course. Can I still join?")
```

```python
assistant.rag("docker")
```

## 3. Add explicit grounding instructions

The custom instructions constrain the assistant to facts found in the retrieved FAQ context. This reduces the risk of supplementing an answer with unsupported general knowledge.

```python
custom_instructions = """
You're a course teaching assistant.
Answer the QUESTION based on the CONTEXT from the FAQ database.
Use only the facts from the CONTEXT when answering the QUESTION.
""".strip()

assistant = RAGBase(
    index=index,
    llm_client=openai_client,
    instructions=custom_instructions,
)
```

```python
assistant.rag("How do I get a certificate?")
```

```python
assistant.rag("Can I still join the course after it started?")
```
